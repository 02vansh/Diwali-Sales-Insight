# Exploratory Data Analysis

# Gender

```python
ax = sns.countplot(x = 'Gender', data = cleandata)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = cleandata.groupby(['Gender'], as_index = False)['Amount'].sum()
sales_gender.head()
```

```python
sns.barplot(x = 'Gender',y = 'Amount', data = sales_gender)
```

From the above exploratory data analysis we can see that most of the buyers are females, and even the purchasing power of female is greater than males.

# Age

```python
age = sns.countplot(x = 'Age Group', data= cleandata)
for bars in age.containers:
    age.bar_label(bars)
```

We can see that most of the customers are from Age Group of 26-35 years.

```python
graph = sns.countplot(x = 'Age Group', data = cleandata, hue = 'Gender')
for bars in graph.containers:
    graph.bar_label(bars)
```

With the help of this visualization we can see that in Age Group of 26-35 years also, Females are high in number as customers.

```python
sales_Age = cleandata.groupby(['Age Group','Gender'],as_index = False)['Amount'].sum()
sales_Age.head(15)
```

```python
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_Age, hue = 'Gender')
sns.set(rc = {'figure.figsize' : (45,5)})
```

From above data visualization we can see that most of the buyers are from Age Group of 26-35 years, Females

# States

```python
sales_state = cleandata.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False)
sales_state.head(20)
```

```python
sns.set(rc={'figure.figsize' : ('25,5')})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
sales_stateamount = cleandata.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sales_stateamount.head(10)
```

```python
sns.set(rc={'figure.figsize' : (25,5)})
sns.barplot(x = 'State', y= 'Amount', data = sales_stateamount)
```

From above exploratory analysis we can see that most of the sales are from Uttar Pradesh, Maharashtra and Karnatak.

# Marital Status

```python
maritalstatus = sns.countplot(x='Marital_Status', data = cleandata)
for bars in maritalstatus.containers:
    maritalstatus.bar_label(bars)
    sns.set(rc={'figure.figsize' : (40,5)})
```

```python
sales_maritalstatus = cleandata.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum()
sales_maritalstatus.head()
```

```python
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_maritalstatus, hue = 'Gender')
sns.set(rc = {'figure.figsize' : (40,5)})
```

From the above graph we can see that most of the buyers are married women and their purchasing power is also high.

# Occupation

```python
axx = sns.countplot(x = 'Occupation', data = cleandata)
sns.set(rc={'figure.figsize':(35,5)})
for bars in axx.containers:
    axx.bar_label(bars)
```

```python
sales_occupation = cleandata.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
sales_occupation.head(25)
```

```python
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

From above graphs we can see that most of the buyers are from IT Sector, Healthcare and Aviation Sector

# Product Category

```python
sns.countplot(x = 'Product_Category', data= cleandata)
```

```python
sales_product = cleandata.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_product.head(20)
```

```python
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
```

```python
sales_order = cleandata.groupby(['Product_Category'], as_index= False)['Orders'].sum().sort_values(by='Orders',ascending= False)
sales_order.head(20)
```

```python
sns.barplot(x='Product_Category',y='Orders',data=sales_order)
```

From above graphs we can see that most sold products are from Food, Clothing and Electronics category

```python
sales_top10 = cleandata.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.barplot(x='Product_ID',y='Orders',data=sales_top10)
```
